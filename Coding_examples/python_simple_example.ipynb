# Using Python for Simple Procedures

This simple notebook shows a simple exercise of downloading a time series, plotting it and getting descriptive statistics.

```python
import yfinance as yf
import pandas as pd
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

## Downloading and Describing Tesla Stocks from Yahoo Finance

```python
tesla = yf.download('TSLA', '2010-01-01', '2023-01-14')
```

```python
tesla
```

```python
plot1 = tesla.Close.plot()

plot1.set(title = "$TSLA Close Price from 2010 to the present",
       xlabel = "Dates", 
       ylabel = "Close Price")

plt.show()
```

```python
tesla.Close.describe()
```

## Elon Musk in the Joe Rogan Podcast

On 2018-09-06, Elon Musk did an interview in the Joe Rogan Podcast and smoked marihuana live, which make the Close Price for the Tesla Stock to fall after appearing in all the finance news. 

MSNBC description of the event: 

"Tesla stock plunged 9 percent on Friday after a bizarre podcast in which CEO Elon Musk smoked weed and sipped whiskey, and two high-profile executives abruptly announced they were leaving the company."

https://www.youtube.com/watch?v=1rS8fFbW57o



```python
tesla_rogan = yf.download('TSLA', '2018-09-05', '2018-09-14')
```

```python
tesla_rogan
```

```python
plot2 = tesla_rogan.Close.plot(marker = 'o')

plot2.set(title = "$TSLA Close Price during the week of his interview with Joe Rogan",
       xlabel = "Dates", 
       ylabel = "Close Price")

plt.show()
```

```python
tesla_rogan.Close.describe()
```

## Multiple Line Plots 

---



Selected from: https://finance.yahoo.com/news/10-stocks-elon-musk-put-110036170.html

```python
tesla = yf.download('TSLA', '2018-01-01', '2023-01-14')
paypal = yf.download('PYPL', '2018-01-01', '2023-01-14')
mod = yf.download('MOD', '2018-05-01', '2023-01-14')
etsy = yf.download('ETSY', '2018-01-01', '2023-01-14')
gamestop = yf.download('OTGLY', '2018-01-01', '2023-01-14')
```

```python
tesla.Close.plot(label = "Tesla")
paypal.Close.plot(label = "Paypal")
mod.Close.plot(label = "Modine Manufacturing")
etsy.Close.plot(label = "Etsy")
gamestop.Close.plot(label = "Gamestop")

plt.xlabel("Close Price")
plt.ylabel("Dates")
plt.title('Plotting Multiple Stocks related to Elon Musk')

plt.legend()

plt.show()
```
